```python
import os
ckpts_folder = "output/Caltech_Pedestrians/eval_ckpts"
```

```python
# AP,AP50,AP75,APs,APm,APl
AP50, bm, bap = dict(), None, 0
for ckpt in sorted(os.listdir(ckpts_folder)):
    with open(os.path.join(ckpts_folder, ckpt, "log.txt")) as f: log = f.read().splitlines()
    if "final" not in ckpt: AP50[int(ckpt.split("_")[1].split(".")[0])] = float(log[-1].split()[-1].split(",")[1])
    else: AP50["final"] = float(log[-1].split()[-1].split(",")[1])
    print(log)
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(15, 5))
plt.xticks(rotation=70)
plt.grid()
plt.plot(AP50.keys(), AP50.values())
```
